```python
import numpy as np
import matplotlib.pyplot as plt
x = 100
learning_rate = 0.1
y = []
for i in range(100):
    x = x - (learning_rate) * (2 * x + 2) # x = x + (-learning_rate)* f'(x)
    y.append(x**2 + 2*x + 5)
    
min_value = x**2 + 2*x + 5
plt.plot(y)
plt.xlabel('Số lần')
plt.ylabel('f(x)')
plt.title('Giá trị f(x) sau số lần thực hiện bước 2')
plt.show()

print(y)
print("Gia tri nho nhat cua ham la: ",min_value)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('data_linear.csv').values # gọi dữ liệu data_linear, chỉ lấy các giá trị
N = data.shape[0] # N là số hàng của data
x = data[:, 0].reshape(-1, 1) # array x có N hàng là giá trị mét vuông của căn nhà x(N,1)
y = data[:, 1].reshape(-1, 1) # array y có N hàng là giá trị của căn nhà đó ứng với mét vuông ở array x.  y(N,1)
plt.scatter(x, y) # vẽ biểu đồ scatter 
plt.xlabel('Mét vuông')
plt.ylabel('Giá')

x = np.hstack((np.ones((N, 1)), x)) # tạo array N hàng và 2 cột trong đó cột đầu tiên là 1 và côt thứ 2 là các phần tử của x

w = np.array([0.,1.]).reshape(-1,1) # tạo arr có 2 hàng và 1 cột có giá trị lần lượt là 0 và 1

numOfIteration = 1000 # số lần huấn luyện
cost = np.zeros((numOfIteration,1)) # tao 1 arr giá trị là 1 có 100 hàng và 1 cột
learning_rate = 0.000001 # tần số huấn luyện
for i in range(1, numOfIteration):
    r = np.dot(x, w) - y # nhân 2 mảng x(N,2) với w(2,1) được 1 mảng (N,1) sau đó trừ đi array y
    cost[i] = 0.5*np.sum(r*r) 
    w[0] -= learning_rate*np.sum(r)
    # correct the shape dimension
    w[1] -= learning_rate*np.sum(np.multiply(r, x[:,1].reshape(-1,1)))
#     print(cost[i])
predict = np.dot(x, w)
plt.plot((x[0][1], x[N-1][1]),(predict[0], predict[N-1]), 'r') 
plt.show()

x1 = 55
y1 = w[0] + w[1] * 50
print(f'Giá nhà cho 50m^2 là : {y1}')
```
